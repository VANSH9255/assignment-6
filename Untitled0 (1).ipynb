```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
x = np.random.rand(100) * 100  # 100 random points for x
y = 0.5 * x + np.random.normal(0, 10, 100)  # Linear relation with some noise

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.6)
plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a random seed for reproducibility
np.random.seed(42)

# Generate a dataset of random numbers (e.g., normally distributed)
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean=0, StdDev=1, 1000 samples

# Create a histogram and a KDE plot
plt.figure(figsize=(12, 6))
sns.histplot(data, bins=30, kde=True, color='skyblue', stat="density", linewidth=0)
plt.title('Distribution of Randomly Generated Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dataset of categories and values
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [23, 45, 12, 67, 30]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(df['Category'], df['Value'], color='cornflowerblue')
plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a random seed for reproducibility
np.random.seed(42)

# Create a dataset with categories and numerical values
categories = ['A', 'B', 'C', 'D']
data = {
    'Category': np.random.choice(categories, size=200),
    'Value': np.random.normal(loc=50, scale=10, size=200)  # Normally distributed values
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='pastel')
plt.title('Distribution of Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset with correlated features
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = 0.5 * x1 + np.random.normal(0, 0.1, n_samples)  # Correlated with x1
x3 = np.random.rand(n_samples)
x4 = 0.7 * x3 + np.random.normal(0, 0.1, n_samples)  # Correlated with x3

# Create a DataFrame
df = pd.DataFrame({
    'Feature1': x1,
    'Feature2': x2,
    'Feature3': x3,
    'Feature4': x4
})

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], alpha=0.6)
ax.set_title('3D Scatter Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate the sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
sales_df = pd.DataFrame(data)

# Step 2: Pivot the data to create a matrix for the heatmap
pivot_df = sales_df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='sum', fill_value=0)

# Step 3: Create the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_df, annot=True, cmap='YlGnBu', fmt='d', linewidths=.5)
plt.title('Sales Variation Across Months and Days')
plt.xlabel('Day of the Month')
plt.ylabel('Month')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate random data
np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.randint(5, 50, size=n)  # Bubble sizes
colors = np.random.choice(['red', 'green', 'blue', 'orange'], size=n)  # Bubble colors

# Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y, 'Sizes': sizes, 'Colors': colors})

# Create a Bokeh scatter plot
output_notebook()
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')
p.circle(df['X'], df['Y'], size=df['Sizes'], color=df['Colors'], alpha=0.6)

show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the data
x = np.linspace(0, 10, 100)  # 100 values from 0 to 10
y = np.sin(x)  # Sine of x

# Create a Bokeh plot
output_notebook()  # Use this if running in a Jupyter Notebook
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')

# Show the plot
show(p)
```

```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate random data
np.random.seed(42)
n = 100
x = np.random.rand(n) * 100  # Random x values
y = np.random.rand(n) * 100  # Random y values
sizes = np.random.randint(5, 50, size=n)  # Random sizes for the markers
colors = np.random.choice(['red', 'green', 'blue', 'orange'], size=n)  # Random colors

# Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y, 'Sizes': sizes, 'Colors': colors})

# Create a Bokeh scatter plot
output_notebook()  # Use this if running in a Jupyter Notebook
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add circles with varying sizes and colors
p.circle(df['X'], df['Y'], size=df['Sizes'], color=df['Colors'], alpha=0.6)

# Show the plot
show(p)
```
